```python
import sklearn
from sklearn import linear_model
```

```python
 x= data.drop('Sales generated per dollar spent',axis = 'columns')
 
```

```python
y= data.drop('Cost for Advertisement', axis = 'columns')
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
a=((reg.score(x,y))*100)
b = round(a,2)
print(f'The efficiency of the model is {b}%')
```
